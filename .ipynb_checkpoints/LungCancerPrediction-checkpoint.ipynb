```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
data_set=pd.read_csv("Lung_Cancer_Prediction/cancer_dataset.csv")
```

```python
data_set
```

```python
selected_data=data_set.loc[:,["Patient Id","Age","Gender","Air Pollution","Alcohol use","Dust Allergy","chronic Lung Disease","Shortness of Breath"]]
```

```python
selected_data.isna().sum()
```

```python
training_x_data=selected_data.loc[:900]
```

```python
training_x_data
```

```python
training_y_data=data_set.loc[:900,["Level"]]
```

```python
training_y_data
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
training_x_data.loc[:,["Patient Id"]]=label_encoder.fit_transform(training_x_data["Patient Id"])
```

```python
training_y_data.Level=label_encoder.fit_transform(training_y_data.Level)
```

```python
training_y_data
```

```python
test_x_data=selected_data.loc[901:]
```

```python
test_x_data
```

```python
test_y_data=data_set.loc[901:,["Level"]]
```

```python
test_y_data.Level=label_encoder.fit_transform(test_y_data.Level)
```

```python
test_y_data.Level.unique()
```

```python
model=LinearRegression()
```

```python
training_x_data
```

```python
test_x_data.loc[:,["Patient Id"]]=label_encoder.fit_transform(test_x_data["Patient Id"])
```

```python
test_x_data
```

```python
model.fit(training_x_data,training_y_data)
```

```python
predicted_data=model.predict(test_x_data)
```

```python
predicted_data=predicted_data.flatten()
```

```python
output=pd.DataFrame({
    "Id":test_x_data["Patient Id"],
    "Actual Level":test_y_data.Level,
    "Predicted level":predicted_data

})
```

```python
output
```

```python
output['Accuracy'] = output.apply(
    lambda row: (row['Actual Level'] - row['Predicted level']) / row['Actual Level']
    if row['Actual Level'] > 0 else row['Actual Level'] - row['Predicted level'],
    axis=1
)
```
