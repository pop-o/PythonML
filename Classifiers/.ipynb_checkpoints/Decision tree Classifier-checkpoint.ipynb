```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
import seaborn as sns
from sklearn import tree
```

```python
classifier=DecisionTreeClassifier()
```

```python
data_set=pd.read_csv("cancer_dataset.csv")
```

```python
data_set
```

```python
x_train=data_set.loc[:899]
```

```python
x_train=x_train.drop(columns=["index","Patient Id","Level"])
```

```python
x_train
```

```python
x_train.isna().sum()
```

```python
y_train=data_set.loc[:899,["Level"]]
```

```python
y_train
```

```python
from sklearn import preprocessing
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
y_train.Level=label_encoder.fit_transform(y_train.Level)
```

```python
y_train
```

```python
x_test=data_set.loc[900:]
```

```python
x_test=x_test.drop(columns=["index","Patient Id","Level"])
```

```python
y_test=data_set.loc[900:,["Level"]]
```

```python
y_test
```

```python
y_test.Level=label_encoder.fit_transform(y_test.Level)
```

```python
y_test
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
y_test
```

```python
y_pred
```

```python
resul=pd.DataFrame(y_test.values,columns=["Test"])
resul["Predicted"]=y_pred
```

```python
resul
```

```python
resul.to_csv("Lung Cancer Prediction using decision treee classifier.csv",index=False)
```

```python
sns.scatterplot(x=y_test.values.flatten(),y=y_pred)
```
